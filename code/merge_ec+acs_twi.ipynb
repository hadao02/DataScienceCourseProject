```python
import pandas as pd
import numpy as np

results_and_twitter = pd.read_csv('/content/drive/MyDrive/tweets/results_and_twitter.csv', index_col=0)
results_and_twitter.drop(['last_name_x','last_name_y', '_m_full', '_m_last'], axis=1, inplace=True)
results_and_twitter['candidate_name_noacc'] = results_and_twitter['candidate_name'].str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8')
results_and_twitter['candidate_name_noacc'] = results_and_twitter['candidate_name_noacc'].str.title()
fec_acs = pd.read_excel('/content/drive/MyDrive/tweets/fec+acs.xlsx', index_col=0)
fec_acs.drop_duplicates(inplace=True)

results_and_twitter_col = list(results_and_twitter.columns.values)
fec_acs_col = list(fec_acs.columns.values)
fec_acs_col.append('last_name_x')
results_and_twitter_col.append('last_name_y')

all_house_candidates = results_and_twitter.loc[results_and_twitter['office_x']=='house']['candidate_name'].unique()
print(results_and_twitter.loc[results_and_twitter['office_x']=='house'].info()) # 1055 hosue candidates
```

```python
##### check for duplicates in fec-acs data
fec_acs = pd.read_excel('/content/drive/MyDrive/tweets/fec+acs.xlsx', index_col=0)
fec_acs.loc[fec_acs.duplicated('Cand_Name')].sort_values('Cand_Name')
```

```python
# merging using full names
def convert_name(name_comma):
  if ',' in name_comma:
    split = name_comma.split(',')
    if len(split) > 2:
      return(name_comma)
    if len(split) == 2:
      return(split[1].strip() + ' ' + split[0].strip())
  else:
    return(name_comma)

def keep_first_last(string):
  split = string.split(' ')
  return(split[0] + ' ' + split[-1])

fec_acs['converted_name'] = fec_acs['Cand_Name'].map(convert_name).map(keep_first_last).str.title()
fec_acs['last_name'] = fec_acs['Cand_Name'].str.split(',').str[0].str.title()

results_and_twitter['last_name'] = results_and_twitter['candidate_name_noacc'].str.split().str[-1]
m_first_last = fec_acs.merge(results_and_twitter, how='outer', left_on='converted_name', right_on='candidate_name_noacc', indicator=True)

final_df = m_first_last.loc[m_first_last['_merge']=='both']
fec_only = m_first_last.loc[m_first_last['_merge']=='left_only']
fec_only = fec_only[fec_acs_col]
twi_only = m_first_last.loc[m_first_last['_merge']=='right_only']
twi_only = twi_only[results_and_twitter_col]

final_df.drop_duplicates(['candidate_name', 'office_x']).info()
final_df.drop_duplicates(['candidate_name']).info() # 858
```

```python
# merging using unique last names
fec_unique_last = fec_only.drop_duplicates(['last_name_x'], keep=False)
fec_last_name_list = fec_unique_last['last_name_x'].unique()
twi_unique_last = twi_only.drop_duplicates(['last_name_y'], keep=False)
twi_last_name_list = twi_unique_last['last_name_y'].unique()

m_last_name = fec_only.merge(twi_only, how='outer', left_on='last_name_x', right_on='last_name_y', sort=True, indicator='_m_last')
m_last_name['unique'] = np.where(m_last_name['last_name_x'].isin(fec_last_name_list) & m_last_name['last_name_y'].isin(twi_last_name_list), True, False)
print(m_last_name['unique'].sum()) # 129

final_df = pd.concat([final_df, m_last_name.loc[m_last_name['unique']==True]])
final_df.drop_duplicates('candidate_name').info() # 986
```

```python
# match by hand
handmatch_df = m_last_name.loc[m_last_name['unique']==False]
temp = handmatch_df[['Cand_Name', 'candidate_name_noacc', 'Cand_Id', 'state', 'office_x', 'district']].loc[handmatch_df['office_x']=='house']
twi_handmatch = handmatch_df.loc[handmatch_df['_m_last'].isin(['right_only', 'both'])][results_and_twitter_col]
fec_handmatch = handmatch_df.loc[handmatch_df['_m_last'].isin(['left_only', 'both'])][fec_acs_col]
twi_handmatch = twi_handmatch.drop_duplicates()
fec_handmatch = fec_handmatch.drop_duplicates()
print(twi_handmatch.info()) # 335 entries
```
